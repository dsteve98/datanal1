# Cluster sampling


```python
from sklearn.impute import SimpleImputer
imp_mode=SimpleImputer(missing_values=numpy.NaN, strategy='most_frequent')
imp_mode.fit(datas)
clean_data = pandas.DataFrame(imp_mode.transform(datas))
clean_data.columns = datas.columns
clean_data.index = datas.index
```

```python
# Encode the categorical features as numbers
import sklearn.preprocessing as preprocessing
def number_encode_features(df):
    result = df.copy()
    encoders = {}
    for column in result.columns:
        if result.dtypes[column] == numpy.object:
            encoders[column] = preprocessing.LabelEncoder()
            result[column] = encoders[column].fit_transform(result[column].fillna('0'))
    return result, encoders
```

```python
#MASIS ERRORRRRRRRRRRRRRR

from sklearn.cluster import KMeans

n_clusters = 2
test_size=0.7
data_without_class,_ = number_encode_features(clean_data)
data_without_class = data_without_class.drop(columns='income')
kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(data_without_class)
selected_index=[]
for i in range(n_clusters):

    C_i = numpy.where(kmeans.labels_ == i)[0].tolist() 
    n_i = len(C_i) # number of points in cluster i
    
    
    sample_i = numpy.random.choice(C_i, int(test_size * n_i)) 
    selected_index.append(sample_i)
print('Index yang terpilih:',selected_index)

train1=data_without_class.iloc[selected_index[0],:]
train2=data_without_class.iloc[selected_index[1],:]

print('Jumlah data dari cluster pertama',len(train1))
print('Jumlah data dari cluster kedua',len(train2))
```
