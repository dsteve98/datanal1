# Noise Reduction

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer

dataset = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',header=None, sep=',\s', engine='python')
dataset = dataset.replace('?', np.NaN)
dataset.columns = ['Age', 'Workclass', 'Fnlwgt', 'Education', 'EducationNum', 'MaritalStatus', 'Occupation', 'Relationship', 'Race','Sex', 'CapitalGain', 'CapitalLoss', 'HoursperWeek', 'NativeCountry', 'Income']

dataset['Workclass'] = dataset['Workclass'].astype('category')
dataset['Education'] = dataset['Education'].astype('category')
dataset['MaritalStatus'] = dataset['MaritalStatus'].astype('category')
dataset['Occupation'] = dataset['Occupation'].astype('category')
dataset['Relationship'] = dataset['Relationship'].astype('category')
dataset['Race'] = dataset['Race'].astype('category')
dataset['Sex'] = dataset['Sex'].astype('category')
dataset['NativeCountry'] = dataset['NativeCountry'].astype('category')
dataset['Income'] = dataset['Income'].astype('category')

dataset.dtypes
cat_columns = dataset.select_dtypes(['category']).columns
cat_columns
dataset[cat_columns] = dataset[cat_columns].apply(lambda x: x.cat.codes)

X = dataset.values
imp_mode = SimpleImputer(missing_values=np.NaN, strategy='most_frequent')
imp_mode.fit(X)
X = imp_mode.transform(X)
df = pd.DataFrame(X)
print(df[:15])
```

## Removing outlier with Z score

```python
#%matplotlib inline
#data2 = df
#data2[0] = pd.to_numeric(data2[0])
#data2[2] = pd.to_numeric(data2[2])
#data2[4] = pd.to_numeric(data2[4])
#data2[10] = pd.to_numeric(data2[10])
#data2[11] = pd.to_numeric(data2[11])
#data2[12] = pd.to_numeric(data2[12])
#data2.boxplot(figsize=(20,3))
#df.head()

#data2 = df.copy()

#data2[1] = data2[3] = data2[5] = data2[6] = data2[7] = data2[8] = data2[9] = data2[13] = data2[14] =  0

Z = (df-df.mean())/df.std()
#Z.fillna(0, inplace=True)

print('Number of rows before discarding outliers = %d' % (Z.shape[0]))

Z2 = Z.loc[((Z > -3).sum(axis=1)==14) & ((Z <= 3).sum(axis=1)==14),:]

for idx in Z2.index.values:
    df.drop([idx], inplace=True)
    
    
print('Number of rows after discarding outliers = %d' % (df.shape[0]))
```

## Initial Accuracy Test

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedShuffleSplit

def calc(data,label):
    X_train, X_test, Y_train, Y_test = train_test_split(data, label, test_size=.2, stratify=label, random_state=0)
    clf_DT = tree.DecisionTreeClassifier(random_state=0)
    clf_DT = clf_DT.fit(X_train, Y_train)

    hasil_testing_DT = clf_DT.predict(X_test)
    accuracy_DT = accuracy_score(Y_test, hasil_testing_DT)
    print(accuracy_DT)

    scores_DT = cross_val_score(clf_DT, data, label, cv=5)
    print(scores_DT)
    print(np.mean(scores_DT),'\n')
```

```python
import pywt
import numpy as np
from sklearn import datasets

data2 = df.copy()
data2.drop([14], axis=1, inplace=True)

data1 = data2.values
target = df[14].tolist()

(cA, cD) = pywt.dwt(data1, 'db1')
print('Approximation result',cA.shape)
print('Detail results',cD.shape)
calc(cA,target)
```

## Feature Selection to 5

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import mutual_info_classif

selector = SelectKBest(mutual_info_classif, k=5).fit_transform(data1, target)
selector = pd.DataFrame(selector)

selectorpriview = selector.copy()
selectorpriview.columns = ['Age', 'EducationNum', 'MaritalStatus', 'Relationship', 'CapitalGain']

print(selectorpriview)

print("Accuracy after feature selection :")
calc(selector,target)
```

## Use PCA to reduce dimension to 2

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA as sklearnPCA

pca = sklearnPCA(n_components=2)
#Xstd_pca = pca.fit_transform(X_std)
X_pca = pca.fit_transform(data1)

plt.scatter(X_pca[:, 0], X_pca[:, 1], marker='o', c=target)
plt.show()
print(X_pca)

print ("Accuracy after pca :")
calc(X_pca,target)
```

# Use LDA to reduce dimension to 2

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA #newer version of sklearn
# from sklearn.lda import LDA #older version of sklearn
lda = LDA(n_components=2)
X_lda = lda.fit(data1, target).transform(data1)

plt.scatter(X_lda[:, 0], X_lda[:, 0], marker='o', c=target)
plt.show()
print(X_lda)

print ("Accuracy after lda :")
calc(X_lda,target)
```
